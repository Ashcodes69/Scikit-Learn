```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    "Customer": ["Ashish", "Riya", "Aman", "Shivam", "Neha", "Mithlesh"],
    "Age": [20, 30, 40, 22, 38, 25],
    "Spendings": [100, 200, 300, 110, 290, 130],
}
df = pd.DataFrame(data)
print(df)
```

```python
X = df[["Age", "Spendings"]]
model = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Group'] = model.fit_predict(X)
print(df)
```

```python
plt.figure(figsize=(6,4))
for group in df['Group'].unique():
  group_data = df[df['Group']==group]
  plt.scatter(group_data['Age'],group_data['Spendings'],label=f'Group {group}')

plt.title("Customer segment (K Means)")
plt.xlabel("Age")
plt.ylabel('Spendings')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
